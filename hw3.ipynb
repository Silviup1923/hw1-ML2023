```python
df = df[['Make','Model','Year','Engine HP','Engine Cylinders','Transmission Type','Vehicle Style','highway MPG','city mpg','MSRP']]
df
```

```python
numerical = ['year','engine_hp','engine_cylinders','highway_mpg','city_mpg']
categorical = ['make','model','transmission_type','vehicle_style']
correlation_matrix = df[numerical].corr()
correlation_matrix[correlation_matrix == correlation_matrix[correlation_matrix <1].max()[-1]]
#highway_mpg and city_mpg are the most correlated features
```

```python
#Making the classification variable binary
df['above_average'] = (df.price > df.price.mean()).astype(int)
df
```

```python
from sklearn.model_selection import train_test_split
df_full_train,df_test = train_test_split(df,test_size = 0.2, random_state=42)
df_train,df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)

y_full_train = df_full_train.above_average.values
y_train = df_train.above_average.values
y_val = df_val.above_average.values
y_test = df_test.above_average.values

len(df_full_train),len(df_test),len(df_val),len(df_train),len(y_full_train),len(y_train),len(y_val),len(y_test)
```

```python
df_train = df_train.drop(['price','above_average'],axis=1)
df_test = df_test.drop(['price','above_average'],axis=1)
df_val = df_val.drop(['price','above_average'],axis=1)
df_test
```

```python
#Feature importance
from sklearn.metrics import mutual_info_score
mutual_info_scores = {}

for c in categorical:
    
    mi = mutual_info_score(df_full_train.above_average,df_full_train[c])
    mutual_info_scores[c] = round(mi,2)

mutual_info_scores = pd.Series(mutual_info_scores)
mutual_info_scores[mutual_info_scores == mutual_info_scores.min()]

#transmission type has the lowest mutual info score
```

```python
#One-hot-encoding
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical + numerical].to_dict(orient = 'records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categorical + numerical].to_dict(orient = 'records')
X_val = dv.transform(val_dict)
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear',C=10, max_iter=1000,random_state=42)
model.fit(X_train,y_train)
```

```python
#Calculating the accuracy of the validation dataset
y_pred = model.predict_proba(X_val)[:,1].round(3)
above_average = (y_pred > 0.5)
(y_val == above_average).mean().round(2)
# 95 % accurracy 
```

```python
#Using the accuracy score of sklearn
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_val,y_pred>0.5).round(2)
acc
```

```python
def new_model(feature):
    
    X = df.drop([feature,'price','above_average'],axis = 1)
    y = df['above_average']
    
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.4, random_state = 42)
    X_test, X_val, y_test, y_val = train_test_split(X_test,y_test, test_size = 0.5, random_state = 42)
    
    train_dict = X_train.to_dict(orient = 'records') 
    dv = DictVectorizer(sparse = False)
    X_train = dv.fit_transform(train_dict)
    
    val_dict = X_val.to_dict(orient='records')
    X_val = dv.transform(val_dict)
    model = LogisticRegression(solver='liblinear',C=10, max_iter=1000,random_state=42)
    model.fit(X_train,y_train)
    y_pred = model.predict_proba(X_val)[:,1]
    acc = accuracy_score(y_val,y_pred >= 0.5)
    return acc

    
    
    
```

```python
scores_without_feature = {}
for col in df.iloc[:,:-2].columns:
    scores_without_feature[col] = (acc - new_model(col))
scores_without_feature
```

```python
df.price = np.log1p(df.price)
df = df.drop(['above_average'],axis=1)
```

```python
#Splitting the dataset
df_train_full, df_test = train_test_split(df,test_size = 0.2,random_state = 42)
df_train,df_val = train_test_split(df_train_full,test_size=0.25,random_state=42)
df_train = df_train.reset_index(drop = True)
df_test = df_test.reset_index(drop = True)
df_val = df_val.reset_index(drop = True)
y_train = df_train.price.values
y_test = df_test.price.values
y_val = df_val.price.values
```

```python
df_train = df_train.drop(['price'],axis = 1)
df_test = df_test.drop(['price'],axis = 1)
df_val = df_val.drop(['price'],axis = 1)
len(df_train),len(df_val),len(df_test)
```

```python
train_dict = df_train.to_dict(orient='records')
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
X_train = dv.transform(train_dict)

val_dict = df_val.to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
#Implementation of Ridge Regression
for a in [0,0.01,0.1,1,10]:
    model = Ridge(alpha=a, solver="sag", random_state =42)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_val)
    score = np.sqrt(mean_squared_error(y_val,y_pred))
    print(a,round(score,9))
```
